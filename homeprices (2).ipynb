```python
def convert_to_marla(row):
    area_size = row['Area Size'] 
    area_type = row['Area Type']
    
    if area_type == "Kanal":
        return area_size * 20 
    elif area_type == "Marla":
        return area_size
    else:
        return None
pf1["area in marla"] = pf1.apply(convert_to_marla, axis=1)
pf1.head()
```

```python
pf2 = pf1.drop(["Area Type","Area Size"],axis="columns")
pf2.head()
```
